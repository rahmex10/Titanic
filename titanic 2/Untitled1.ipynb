```python
def title(x):
    if x == 'Mr.':
        return 14
    elif x == 'Mrs.':
        return 13
    elif x == 'Miss.':
        return 12
    elif x == 'Master.':
        return 11
    elif x == 'Dr.':
        return 10
    elif x == 'Rev.':
        return 9
    elif x == 'y':
        return 8
    elif x == 'Planke,':
        return 7
    elif x == 'Impe':
        return 6
    elif x == 'Gordon,':
        return 5
    elif x == 'Col.':
        return 4
    elif x == 'Mile.':
        return 3
    elif x == 'Major.':
        return 2
    else:
        return 1
```

```python
df['Title'] = df['Title'].apply(title)
```

```python
def age(x):
    Age = x[0]
    Pclass = x[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return np.random.randint(38, 50)
        elif Pclass == 2:
            return np.random.randint(29, 37)
        else:
            return np.random.randint(19, 26)
    else:
        return Age
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(age, axis = 1)
```
